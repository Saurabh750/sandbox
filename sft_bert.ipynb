```python
tokenized_datasets = tokenized_datasets.remove_columns(["text"])
```

```python
tokenized_datasets = tokenized_datasets.rename_column("label", "labels")
```

```python
tokenized_datasets.set_format("torch")
```

```python
small_train_dataset = tokenized_datasets["train"].shuffle(seed=42).select(range(10000))
small_eval_dataset = tokenized_datasets["test"].shuffle(seed=42).select(range(1000))
```

```python
from torch.utils.data import DataLoader

train_dataloader = DataLoader(small_train_dataset, shuffle=True, batch_size=16)
eval_dataloader = DataLoader(small_eval_dataset, batch_size=16)
```

```python
print(len(train_dataloader))  # Number of batches run in a single epoch
```

```python
from transformers import AutoModelForSequenceClassification

model = AutoModelForSequenceClassification.from_pretrained("google-bert/bert-base-cased", num_labels=5)
```

```python
from torch.optim import AdamW

optimizer = AdamW(model.parameters(), lr=5e-5)
```

```python
from transformers import get_scheduler

num_epochs = 3
num_training_steps = num_epochs * len(train_dataloader)
lr_scheduler = get_scheduler(
    name="linear", optimizer=optimizer, num_warmup_steps=0, num_training_steps=num_training_steps
)
# "name" takes linear, constant, cosine, cosine_with_restarts
# "linear": LR linearly decreases from initial value to zero 
# "constant": LR remains constant
# "cosine": follows a cosine decay from the initial value to a minimum value
# "cosine_with_restarts": complex version of cosine annealing where the learning rate 
#                         decays following a cosine curve but with periodic restarts.

# num_warmup_steps: The number of steps during which the learning rate will increase linearly 
#                   from 0 to the initial learning rate, before decaying according to the 
#                   specified scheduler
# Used so that model can gradually adapt to the task without large jumps in the learning rate.
```

```python
import torch
from accelerate.test_utils.testing import get_backend

device, _, _ = get_backend() # automatically detects the underlying device type (CUDA, CPU, XPU, MPS, etc.)
print(device)
model.to(device)
```

```python
for batch in train_dataloader:
    for k, v in batch.items():
        print(k, v)
```

```python
from tqdm.auto import tqdm

progress_bar = tqdm(range(num_training_steps))

model.train()
for epoch in range(num_epochs):
    for batch in train_dataloader:
        batch = {k: v.to(device) for k, v in batch.items()}
        outputs = model(**batch)
        loss = outputs.loss
        loss.backward()

        optimizer.step()
        lr_scheduler.step()
        optimizer.zero_grad()
        progress_bar.update(1)
```

```python
output_dir = './saved_model_1'  
model.save_pretrained(output_dir)
```

```python
import evaluate

metric = evaluate.load("accuracy")
model.eval()
for batch in eval_dataloader:
    batch = {k: v.to(device) for k, v in batch.items()}
    with torch.no_grad():
        outputs = model(**batch)

    logits = outputs.logits
    predictions = torch.argmax(logits, dim=-1)
    metric.add_batch(predictions=predictions, references=batch["labels"])

metric.compute()
```

# Model training and evaluation in same step. I will save only that model with best accuracy. Accuracy will be calculated after each epoch

```python
from datasets import load_dataset

dataset = load_dataset("yelp_review_full")
```

```python
from transformers import AutoTokenizer

tokenizer = AutoTokenizer.from_pretrained("google-bert/bert-base-cased")


def tokenize_function(examples):
    return tokenizer(examples["text"], padding="max_length", truncation=True)


tokenized_datasets = dataset.map(tokenize_function, batched=True)
```

```python
tokenized_datasets
```

```python
train_data = tokenized_datasets['train']
train_data = train_data.shuffle(seed=42)
```

```python
small_train_dataset = train_data.select(range(10000))  
small_eval_dataset = train_data.select(range(10000, 11000))
```

```python
small_test_dataset = tokenized_datasets["test"].shuffle(seed=42).select(range(1000))
```

```python
from torch.utils.data import DataLoader

train_dataloader = DataLoader(small_train_dataset, shuffle=True, batch_size=16)
eval_dataloader = DataLoader(small_eval_dataset, batch_size=16)
test_dataloader = DataLoader(small_test_dataset, batch_size=16)
```

```python
from transformers import AutoModelForSequenceClassification

model = AutoModelForSequenceClassification.from_pretrained("google-bert/bert-base-cased", num_labels=5)
```

```python
from torch.optim import AdamW

optimizer = AdamW(model.parameters(), lr=5e-5)
```

```python
from transformers import get_scheduler

num_epochs = 10
num_training_steps = num_epochs * len(train_dataloader)
lr_scheduler = get_scheduler(
    name="linear", optimizer=optimizer, num_warmup_steps=0, num_training_steps=num_training_steps
)
```

```python
import torch
from accelerate.test_utils.testing import get_backend

device, _, _ = get_backend() # automatically detects the underlying device type (CUDA, CPU, XPU, MPS, etc.)
print(device)
model.to(device)
```

```python
from tqdm.auto import tqdm
import evaluate

metric = evaluate.load("accuracy")

progress_bar = tqdm(range(num_training_steps))

best_accuracy = -1

for epoch in range(num_epochs):
    model.train()
    for batch in train_dataloader:
        batch = {k: v.to(device) for k, v in batch.items()}
        outputs = model(**batch)
        loss = outputs.loss
        loss.backward()

        optimizer.step()
        lr_scheduler.step()
        optimizer.zero_grad()
        progress_bar.update(1)
        
    model.eval()
    for batch in eval_dataloader:
        batch = {k: v.to(device) for k, v in batch.items()}
        with torch.no_grad():
            outputs = model(**batch)
    
        logits = outputs.logits
        predictions = torch.argmax(logits, dim=-1)
        metric.add_batch(predictions=predictions, references=batch["labels"])
    
    eval_accuracy = metric.compute()
    print("Evaluation Accuracy: ", eval_accuracy)
    if eval_accuracy['accuracy'] > best_accuracy:
        best_accuracy = eval_accuracy['accuracy']
        print(f"Best Accuracy found in epoch {epoch}", best_accuracy)
        output_dir = './saved_model_2_{epoch}'  
        model.save_pretrained(f"./saved_model_2_{epoch}")
        
```
